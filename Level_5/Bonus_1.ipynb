```python
df1 = pd.read_csv('datasets/House_Price_dataset.csv')
df2 = pd.read_csv('datasets/Titanic.csv')
```

```python
def column_mean(df):
    logging.info('The process has started!')
    col = []
    for column in df.columns:
        if df[column].dtypes == 'int' or df[column].dtype == 'float':
            col.append(column)
            
    mean = df[col].mean()
    logging.info('The process has ended!')
    print('Average :\n',mean)
    
    return mean
```

```python
df1_mean = column_mean
df2_mean = column_mean
```

```python
th1 = threading.Thread(target=column_mean, args = (df1,))
th2 = threading.Thread(target=column_mean, args = (df2,))

th2.start()
th1.start()

th2.join()
th1.join()
```
