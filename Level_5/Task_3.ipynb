```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('datasets\Titanic.csv')
```

```python
df.head(7)
```

```python
summary_stats = df[['Age','Fare','Pclass','Survived']].describe()
summary_stats
```

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()
survival_by_class.head()
```

```python
print('Summary Statistics:')
summary_stats
```

```python
sns.barplot(survival_by_class)
plt.title('Probability of survived based on Passenger class')
plt.show()
```
