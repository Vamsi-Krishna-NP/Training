```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
arr = np.random.randint(1,500,100)
```

```python
print("Length : ",len(arr),"\n",arr)
```

```python
sns.boxplot(data=arr)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Boxplot on Random values')
```
