```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.pareto(a = 2, size= 1000)
```

```python
plt.hist(data, bins=50, density=True, log=True, edgecolor = 'black')
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Pareto Distribution')
plt.show()
```

```python
sorted_data = np.sort(data)[::-1]
top20 = sorted_data[:200]
```

```python
print(len(sorted_data))
```

```python
tot_sum = np.sum(sorted_data)
top20_sum = np.sum(top20)
```

```python
print(f"Total Sum: {tot_sum}")
print(f"Top 20% Sum: {top20_sum}")
print(f"Percentage of Total Sum from Top 20%: {top20_sum / tot_sum * 100:.2f}%")
```
