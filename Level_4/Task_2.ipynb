```python
from PyPDF2 import PdfReader,PdfWriter
import PyPDF2 as pdf
```

```python
file_pdf = PdfReader('datasets\sample.pdf')
out_pdf = PdfWriter()
```

```python
len(file_pdf.pages)
```

```python
page = file_pdf.pages[0]
print(page.extract_text())
```

```python
out_pdf.add_page(page)
```

```python
out_pdf.encrypt("password")
with open("password.pdf", "wb") as f:
    out_pdf.write(f)
```
