```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
```

```python
data = pd.read_csv('dataset/USA Housing Dataset.csv')
```

```python
data.head()
```

```python
corr_pearson, p_value_pearson = pearsonr(data['price'], data['sqft_living'])
print(f"Pearson correlation: {corr_pearson}, P-value: {p_value_pearson}")
```

```python
corr_spearman, p_value_spearman = spearmanr(data['price'], data['sqft_living'])
print(f"Spearman correlation: {corr_spearman}, P-value: {p_value_spearman}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sqft_living', y='price', data=data)
plt.title('Price vs sqft_living')
plt.xlabel('sqft_living')
plt.ylabel('Price')
plt.show()
```

```python
numerical_features = data.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_features.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
